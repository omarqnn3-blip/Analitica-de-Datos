```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Datos_Facturación.csv')
df
```

```python
filtro1 = df[(df['CVE_CLPV']>= 1000) & (df['CVE_CLPV'] <= 2000)]
filtro1
```

```python
filtro1.to_csv('Filtro CVE.csv')
```

```python
filtro2 = df[~(df['CVE_VEND']== 4) & ~(df['CVE_VEND'] ==5 )]
filtro2
```

```python
df=filtro2.groupby(['CVE_VEND'])['CVE_VEND'].count()
df
```

```python
filtro2.to_csv('Filtro CVE_VEND.csv')
```

```python
filtro3 = df[df['FECHA_ENT'] == "2023-02-28 00:00:00"]
filtro3
```
